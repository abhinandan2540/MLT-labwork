```python
# experimenting on the effect of cognitive load on sleeping hourss
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
import seaborn as sns

csv_data=pd.read_csv('ai_dev_productivity.csv')
# print(csv_data.head())
# train model on 80% data, test on 20% data
x_train=csv_data['sleep_hours'][:80]
y_train=csv_data['cognitive_load'][:80]

# sleep hours mean
# x_train_mean=math.ceil(np.mean(x_train))
# print(x_train_mean)

# # cognitive overload mean
# y_train_mean=math.ceil(np.mean(y_train))
# print(y_train_mean)

# x_train_mean = np.mean(x_train)
# print(x_train_mean)

# # cognitive overload mean
# y_train_mean = np.mean(y_train)
# print(y_train_mean)

# # sleep hours median
x_train_median=np.median(x_train)
print(x_train_median)

# # cognitive overload median
y_train_median=np.median(y_train)
print(y_train_median)

# in that case, mean/median/mode dosen't give the actual nature of data. 
plt.scatter(x_train,y_train, alpha=0.5,color='blue')
plt.axvline(x=x_train_median,color='red')
plt.axhline(y=y_train_median,color='red')
plt.xlabel('sleep hours')
plt.ylabel('cognitive overload')
plt.show()
```
