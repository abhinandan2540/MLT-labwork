# INFORMATION REGARDING ILPD
# Total Features : 10
# Target : Dataset (1:Patient has liver disease, 2: patient does not have liver disease)

# Age (numeric)
# Gender (Male/Female â†’ categorical)
# Total_Bilirubin (numeric)
# Direct_Bilirubin (numeric)
# Alkaline_Phosphotase (numeric)
# Alamine_Aminotransferase (ALT) (numeric)
# Aspartate_Aminotransferase (AST) (numeric)
# Total_Proteins (numeric)
# Albumin (numeric)
# Albumin_and_Globulin_Ratio (numeric)


```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn import tree
import matplotlib.pyplot as plt

encoder=LabelEncoder()
csv=pd.read_csv('ILPD_headers.csv')
csv['Gender'] = encoder.fit_transform(csv['Gender'])

# x,y
x=csv.values[:,0:10]
y=csv.values[:,10]

# training testing data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=100)

# gini cofficient
cofficient_gini = DecisionTreeClassifier(criterion='gini',max_depth=None,splitter='best',min_samples_leaf=9,random_state=100)
cofficient_gini.fit(x_train,y_train)

y_pred=cofficient_gini.predict(x_test)

# model evaluation
print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test, y_pred)*100)
print(classification_report(y_test, y_pred))


# for plotting
feature_names = ["Age", "Gender","Total_Bilirubin","Direct_Bilirubin","Alkaline_Phosphotase","Alamine_Aminotransferase","Aspartate_Aminotransferase", "Total_Proteins","Albumin", "Albumin_and_Globulin_Ratio"]
class_names = ["Liver Disease", "No Liver Disease"]

plt.figure(figsize=(100, 100))
tree.plot_tree(cofficient_gini, filled=True, feature_names=feature_names,
               class_names=class_names, rounded=True)
plt.show()
```
