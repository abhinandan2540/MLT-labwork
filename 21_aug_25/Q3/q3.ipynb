```python
import pandas as pd

csv_data=pd.read_csv('titanic_csv.csv')
# print(csv_data.head())
```

(a) Histogram for a numerical variable (e.g., Age or Fare) to visualize its distribution

```python
import numpy as np
# in our age, we have lot's of unfilled data
# removing unfilled data
# dropna: clear all the null values from the dataset
new_csv_data=csv_data.dropna()
# print(new_csv_data.info())
# print(new_csv_data)

x_age=np.array(new_csv_data['Age'])
# print(x_age)

x_fare=np.array(new_csv_data['Fare'])

import matplotlib.pyplot as plt
# age histogram
plt.hist(x_age,bins=30,color='skyblue', edgecolor='black')
plt.xlabel('Age')
plt.ylabel("Frequency")
plt.title('Age Histogram Plot')
plt.show()
```

```python
# fare histogram
plt.hist(x_fare, bins=30, color='skyblue', edgecolor='black')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.title('Fare Histogram Plot')
plt.show()
```

(b) Pie chart for a categorical feature (e.g., Sex or Embarked) to show composition.

```python
# for pie char of sex
# we need the frequency of the sex's

sex_frequency=csv_data['Sex'].value_counts()
print(sex_frequency)
sex_frequency_arr=np.array(sex_frequency)
# print(arr)

# plotting sex frequency into pie chart
sex_labels=['Male','Female']
sex_colors=['skyblue','pink']
max_explode=[0.1,0]
plt.pie(sex_frequency_arr,labels=sex_labels,colors=sex_colors,explode=max_explode,shadow=True)
plt.legend()
plt.show()
```

```python
# pie chart plotting for Embarked
# count frequency for embarked

embarked_frequency=csv_data['Embarked'].value_counts()
print(embarked_frequency)

embarked_frequency_arr=np.array(embarked_frequency)
# print(embarked_frequency_arr).

embarked_labels=['S','C','Q']
plt.pie(embarked_frequency_arr,labels=embarked_labels,shadow=True)
plt.legend()
plt.show()
```

(c)Box plot for a numerical variable (e.g., Age or Fare) to detect potential outliers.

```python
boxplot_vars=[x_age,x_fare]

plt.boxplot(boxplot_vars)
plt.Figure()
plt.show()
```

(d)Correlation heatmap (for numerical columns) to visualize pairwise correlations.

```python
# for correlation headmap
# we need to install seaborn
# pip install seaborn

import seaborn as sns
# compute correlation matrix
corr_matrix=csv_data.corr(numeric_only=True)
print(corr_matrix)

# plotting heatmap using sns
sns.heatmap(corr_matrix,cmap="YlGnBu", annot=True)
plt.show()
```
