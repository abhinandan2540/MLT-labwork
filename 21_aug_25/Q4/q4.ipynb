(4)Feature Encoding Techniques (Label Encoding, One Hot Encoding, and Dummy Encoding)

* Encoding is a process of converting non numeric value into a numeric value for better training and testing of
ML model and algorithm.

Mainly there are 3 types of encoding

1. Label encoding
---------------------------
It assign each category in a feature column a unique integer value
let say we have a column 
color=['red','blue','green']
after encoding..
color=[2,0,1]

label encoding is simple, and no increase in dimentionality
label encoding implies an 'ordinal relationship' (red>green>blue), which might be different that real life scenerio


2. One Hot encoding
-------------------------------------
In this encoding, creates binary colums for each category, with 0,1 0: absent, 1: present
taking the same example
color=['red','blue','green']

after one hot encoding
color_red   color_blue  color_green
    1           0           0
    0           1           1
    0           0           1

In one hot encoding, there is no 'ordinal relationship'  
but it increases the dimentionality


3. Dummy encoding
-------------------------------------
Similar to one hot encoding, but it avoids one category to avoid dummy variable trap

so in the previous example, if we apply dummy encoding
we'll get
Color_Red	Color_Blue
    1	        0
    0	        1
    0           0
    0           1

color_green will be removed

Dummy encoding good as it avoids multicollinarity
but it still increases dimentionality.

(a) Apply Label Encoding to a suitable categorical column (e.g., Embarked or Sex).
 Explain what the resulting numeric labels represent.

```python
import pandas as pd
csv_data=pd.read_csv('titanic_csv.csv')
# print(csv_data.head())

# applying label encoding on 'sex' 
csv_data['label_sex']=pd.factorize(csv_data['Sex'])[0]
# [0] usage because pd.factorize return set of two tuples
# codes, and values of the particular featues
# we need to select the codes, so we are using [0]


# applying label encoding on 'embarked' 
csv_data['label_embarked']=pd.factorize(csv_data['Embarked'])[0]
print(csv_data)
```

(b) Apply One-Hot Encoding to the same column(s). Show how k new binary columns
 are created for k categories. Comment on the representation and redundancy in this
 encoding scheme.

```python

onehotencode_data=pd.get_dummies(csv_data,columns=['Sex','Embarked'],dtype=int)
print(onehotencode_data)
```

(c)Apply Dummy Encoding (a variant of One-Hot Encoding). Show how k−1 columns
 are created for k categories by dropping one reference category. Explain how the
 dropped category is represented implicitly.

```python
dummyencode_data=pd.get_dummies(csv_data,columns=['Embarked'],dtype=int, drop_first=True)
print(dummyencode_data)
```
