Multiple linear regression with scikit learn with model evalutation

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
# using sklearn.linear regress
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error,mean_squared_error,root_mean_squared_error

csv_data=pd.read_csv('Housing.csv')
# print(csv_data.head())

x=np.array(csv_data[['area','bedrooms','bathrooms','stories','parking']])
y=np.array(csv_data['price'])

# 70% training , 30% testing
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=120)

MLR_model=LinearRegression()
# training model with training data
MLR_model.fit(x_train,y_train)

# model predictions
Model_Predictions=MLR_model.predict(x_test)
plt.scatter(y_test,Model_Predictions,color='blue',alpha=0.5,label='Price DP')
plt.plot(y_test,y_test,'r--',lw=3, label='Regression Line')
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.legend()
plt.title('Multiple Linear Regression (LinearRegress())')
plt.show()

# print(f'Actual price: {y_test}')
# print(f'Predicted Price :{Model_Predictions}')

# model evaluation
model_MAE=mean_absolute_error(y_test,Model_Predictions)
model_MSE=mean_squared_error(y_test,Model_Predictions)
model_RMSE=root_mean_squared_error(y_test,Model_Predictions)

# print(model_RMSE)
# print(model_MSE)
# print(model_MAE)
```
