```python
# implementing Linear Regression Model using Scikit learn

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
csv_data=pd.read_csv('Housing.csv')
# print(csv_data.head())
# features for linear regression
x=np.array(csv_data['area']).reshape(-1,1)
y=np.array(csv_data['price']).reshape(-1,1)

# training and testing data using train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=101)
# model training
try:
    my_model=LinearRegression()
    my_model.fit(x_train,y_train)

    # for pridicting prices
    model_prediction=my_model.predict(x_test)
    plt.scatter(y_test, model_prediction, color='blue',alpha=0.5,label='Predicted Price')
    plt.scatter(y_test,y_test, color='red',alpha=0.4, label='Actual Price')
    # plt.plot(x_test,y_test)
    plt.title('Simple Linear Regression (LinearRegress())')
    plt.xlabel('Actual Price')
    plt.ylabel('Predicted Price')
    plt.legend()
    plt.show()
except Exception as e:
    print(e)


# Assumptions of linear regression
# 1. Linearity (a linear relationship betweeen variable)
# 2. erros are not correlated
# 3. Homoscedicity (constant variance of error)
# 4. Normality (errors are normally distributed)
# 5. No multicollinearity (independent variables are not much correlated with each other)

# mse=mean_squared_error(y_test,model_prediction)
# mae=mean_absolute_error(y_test,model_prediction)
# print(mse)
# print(mae)
```
