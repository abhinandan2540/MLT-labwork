```python
# This code snippet will discuss how StandardScaler() in sklearn actually works
# standardscaler : do z score standardizing 
# x`=x-mean(x)/std(x)

# on single feature
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

csv = pd.read_csv('pmt.csv')

x = np.array(
    csv['parent_mental_health'])
y = np.array(csv['mental_trauma'])

# splitting data for train_test
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=43,test_size=0.2)

class MyStandardScaler:
    def __init__(self):
        self.mean_=None
        self.std_=None

    def fit(self,x):
        # axis=0 if multiple feature x is input
        self.mean_=np.mean(x,axis=0)
        self.std_=np.std(x, axis=0)
        return self

    def transform(self,x):
        return (x-self.mean_)/(self.std_)

    def fit_transform(self,x):
        return self.fit(x).transform(x)



scaler=MyStandardScaler()
# x_scaled=scaler.fit(x_train)
# print(x_scaled.mean_)
# print(x_scaled.std_)


x_scaled=scaler.fit_transform(x_train)
print(x_scaled)

# 1. x_train first go to fit(x_train) : calculate mean_, std_, then
# 2. go to transform(x_train): with calculated mean_, std_.
# 3. then trasnform(x_test): transform x_test with calculated mean_, std_

# print(scaler.mean_)
# print(scaler.std_)

# transforming x_test data on mean_, std_ of x_train
X_test=scaler.transform(x_test)
print(X_test)

```
