```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay

iris=load_iris()

X=iris.data
y=iris.target

X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.3, # 70% training, 30% testing
                                                   random_state=42 # reporducibility
                                                   )
```

*2(a) Gaussian_NB*

```python
gaussian_nb=GaussianNB()
gaussian_nb.fit(X_train, y_train)
```

```python
y_gaussian_pred=gaussian_nb.predict(X_test)
```

```python
# accuracy
gaussian_accuracy=accuracy_score(y_true=y_test, y_pred=y_gaussian_pred)
print(gaussian_accuracy)
```

```python
# confusion matrix
import matplotlib.pyplot as plt
cm_gaussian=confusion_matrix(y_true=y_test, y_pred=y_gaussian_pred,labels=gaussian_nb.classes_)
conf_matrix_gaussian=ConfusionMatrixDisplay(confusion_matrix=cm_gaussian,display_labels=iris.target_names)
conf_matrix_gaussian.plot()
plt.show()
```

*2(b) Multinomial_NB*

```python
from sklearn.naive_bayes import MultinomialNB

multinomial_nb=MultinomialNB()
multinomial_nb.fit(X_train, y_train)
```

```python
y_multinomial_pred=multinomial_nb.predict(X_test)
```

```python
# multinomial_accuracy
multinomial_accuracy=accuracy_score(y_true=y_test, y_pred=y_multinomial_pred)
print(multinomial_accuracy)
```

```python
cm_multinomial=confusion_matrix(y_true=y_test, y_pred=y_multinomial_pred, labels=multinomial_nb.classes_)
conf_matrix_multinomial=ConfusionMatrixDisplay(confusion_matrix=cm_multinomial, display_labels=iris.target_names)
conf_matrix_multinomial.plot()
plt.show()
```

*F1 score of both models*

```python
from sklearn.metrics import f1_score

# Gaussian NB `f1_score`
Gaussian_f1=f1_score(y_true=y_test, y_pred=y_gaussian_pred, average='weighted')
print(Gaussian_f1)
```

```python
# Multinomial NB `f1_score`
Multinomial_f1=f1_score(y_true=y_test, y_pred=y_multinomial_pred, average='weighted')
print(Multinomial_f1)
```
