```python
# polynomial regression OLS
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error,r2_score
csv=pd.read_csv('poly_data.csv')
# print(csv.head())

x=csv['X']
y=csv['y']
n=len(y)

# for 3 order polynomial regression
# X shape: nxm
X=np.c_[np.ones((n,1)),x,x**2,x**3]
xT=np.transpose(X)
xTx=np.dot(xT,X)

# # parameter 1
xTx_inverse=np.linalg.inv(xTx)

# # parameter 2
xTy=np.dot(xT,y)

# # beta cofficient = dot product (parameter 1, parameter 2)
beta_cofficient=np.dot(xTx_inverse,xTy)

# prediction
y_prediction=np.dot(X,beta_cofficient)

# mean square error
mse=mean_squared_error(y_prediction,y)
r2=r2_score(y_prediction,y)
print(mse)
print(r2)


plt.scatter(x,y,color='blue',alpha=0.4)
plt.plot(x,y_prediction,color='red',label='Regression Line')
plt.xlabel('X')
plt.ylabel("Y")
plt.legend()
plt.title('Polynomial Regression (OLS)')
plt.show()
```
