(b) Perform Multiple Linear Regression (all features â†’ price).

```python
# multiple linear regression using scikit learn
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
csv_data=pd.read_csv('Housing.csv')
# print(csv_data.head())

# area, bedrooms, bathrooms, stories, parking
x=np.array(csv_data[['area','bedrooms','bathrooms','stories','parking']])
y=np.array(csv_data['price'])

# # using sklearn.linearRegress for multiple regression
regression_model=linear_model.LinearRegression()
regression_model.fit(x,y)

def PredictionModel(x):
    return regression_model.predict(x)

try:
    modelOutput=PredictionModel(x)

    plt.scatter(y,modelOutput,color='red',label='predicted price')
    plt.scatter(y,y,alpha=0.2,color='blue',label='actual price')
    plt.xlabel('Actual Price')
    plt.ylabel("predicted Price")
    plt.legend()
    plt.title('Multiple Linear Regression using Lineargression()')
    plt.show()
except Exception as e:
    print(e)
```

(a) Perform Simple Linear Regression (area â†’ price) using  sklearn.linear_model.LinearRegression.

```python
# simple linear regression using scikit learn

import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
csv_data = pd.read_csv('Housing.csv')
# area : feature, price: target
x=np.array(csv_data['area'])
y=np.array(csv_data['price'])
x_area=x.reshape(-1,1)
y_price=y.reshape(-1,1)

regression_model=linear_model.LinearRegression()
regression_model.fit(x_area,y_price)

def predictionModel(x):
    return regression_model.predict(x)

plt.scatter(x_area,y_price,label='Actual Price')
plt.plot(x_area,predictionModel(x_area),color='red',label='Predicted Price')
plt.xlabel('area')
plt.ylabel('price')
plt.title('Simple Linear Regression using LinearRegress()')
plt.legend()
plt.show()
```
