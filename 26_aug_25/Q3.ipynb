(b) Perform Multiple Linear Regression (all features â†’ price) with scikit learn

```python
# multiple linear regression using scikit learn
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
csv_data=pd.read_csv('Housing.csv')
# print(csv_data.head())

# area, bedrooms, bathrooms, stories, parking
x=np.array(csv_data[['area','bedrooms','bathrooms','stories','parking']])
y=np.array(csv_data['price'])


# using scikitlearn.LinearRegress for multiple regression
regression_model = linear_model.LinearRegression()
regression_model.fit(x, y)


def PredictionModel_MLR(x):
    return regression_model.predict(x)

y_predicted=PredictionModel_MLR(x)
try:
    plt.scatter(y,y_predicted, color='blue',alpha=0.5,label='Price DP')
    plt.plot(y,y,'r--',lw=3, label='Regression Line')
    plt.xlabel('Actual Price')
    plt.ylabel("predicted Price")
    plt.legend()
    plt.title('Multiple Linear Regression (Lineargression())')
    plt.show()
except Exception as e:
    print(e)
```

(a) Perform Simple Linear Regression (area â†’ price) using  sklearn.linear_model.LinearRegression.

```python
# simple linear regression using scikit learn

import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
csv_data = pd.read_csv('Housing.csv')
# area : feature, price: target
x=np.array(csv_data['area'])
y=np.array(csv_data['price'])
x_area=x.reshape(-1,1)
y_price=y.reshape(-1,1)

regression_model=linear_model.LinearRegression()
regression_model.fit(x_area,y_price)

def predictionModel_SLR(x):
    return regression_model.predict(x)

plt.scatter(x_area,y_price,color='blue',alpha=0.5,label='Actual DP')
plt.plot(x_area,predictionModel_SLR(x_area),'r--',lw=3,label='Predicted DP')
plt.xlabel('area')
plt.ylabel('price')
plt.title('Simple Linear Regression (LinearRegress())')
plt.legend()
plt.show()
```
